### Import Modules

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None
%matplotlib inline
```

### Dataset

Dataset used in this notebook are from https://www.megadice.com

### Leader Board Winners History

#### Read in the data

```python
leaders = pd.read_json('https://session.megadice.com/globalstats/leaderBoardWinners/')
```

#### Explore data

```python
leaders.head()
```

#### Cleanup and manipulate data

```python
del leaders['queryTimeInSeconds']
leaders.head()
```

```python
leaders.reset_index(inplace=True)
```

```python
leaders.head()
```

```python
leaders2 = leaders.set_index('index').T
```

```python
leaders2.head()
```

```python
leaders2.reset_index(inplace=True)
```

```python
leaders2.head()
```

```python
leaders2["Date"] = pd.to_datetime(leaders2["index"], format="%Y-%m-%d")
```

```python
leaders2.info()
```

```python
leaders2.index = leaders2['Date']
del leaders2['Date']
del leaders2['index']
```

```python
leaders2.head()
```

#### Visualize Data

```python
leaders2.plot()
plt.show()
```
